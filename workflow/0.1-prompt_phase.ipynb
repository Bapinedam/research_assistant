```python


# The research aim is important because the LLM needs to understand the overall context

# and purpose of the analysis. This helps it stay aligned with the research objectives

# and apply the codes in a relevant way.


RESEARCH_AIM =  """

                Our research investigates how students experience feedback when engaging with AI 
                compared with teacher feedback. We aim to understand how students perceive 
                differences between AI and teacher feedback in order to improve feedback practices 
                in the age of AI. This study contributes to pedagogy, technology development, and 
                policy by mapping which aspects of feedback students value, and where they see AI 
                as better, worse, or similar to human teachers.

                """


# The coding framework (categories and subcategories) is critical because it defines
# the deductive structure the LLM should follow. This is the 'skeleton' of the codebook.

# Each code should have a clear definition and hierarchical organization.



# The coding framework (categories and subcategories) is critical because it defines

# the deductive structure the LLM should follow. This is the 'skeleton' of the codebook.

# Each code should have a clear definition and hierarchical organization.

CODING_FRAMEWORK =  """

# Reason for using GenAI:

    1. Processes
        * Access
        a. Ease: Is easily accessible or available.
            * Synonyms: Accessible, available, convenient.
            * Quote: “It’s easier than trying to reach professors and teachers”
        b. Speed: Is immediate, instant or fast.
            * Synonyms: Time-efficient.
            * Quotes: “the speed in which AI can give responses is much faster”, "It helped save time spent on the “less important” aspects of assignments"
        c. Volume: Can be asked a lot of questions and therefore provides more volume of information.
            * Quotes: “much more accessible meaning I could ask a lot more questions”, “Teacher gives very little feedback”
        
        * Timing
        d. Before submission: Provides me with information during the production of an assignment task.
            * Quote: “genai feedback is instant and available prior to assessment submissions, which helps improve writing as it provides feedback on drafting for final submission.”
        e. After submission: Provides me with information after the submission of an assignment task.
            * Quotes: “teacher feedback is helpful in hindsight”, “teacher feedback however is provided after assessment is submitted which is helpful in how the writing COULD have been improved, but is not applicable help since the assessment is already completed.”
        
        * Effort
        f. Off-loading: Off-loads cognitive effort or does work for me.
            * Note: If comment is about speed, making something faster, etc.. then it is coded under "Speed".
            * Quotes: “while teachers are useful for explain content and answering questions, they do not help you complete the assignment to the best of your ability”, "give me some opinion in language as english is not my mother langauage", "It makes my work easier", "Some easy problems can be solved by AI"
    2. Sense-making
        * Understanding
        a. Helps me understand: Helps me understand/confuses me.
            * Quote: “it has helped me organise my thoughts and cohesively express them and understand concepts/events at a basic level”
        * Reflection
        b. Helps me reflect: Helps me reflect.
            * Quote: “AI made me more reflective”
        * Progress
        c. Helps me know how I’m going: Helps me know how I’m going.
            * Quote: “It helped me understand how I was fairing”
    3. Information
        * Quality
        a. Specificity: Provides specific information.
            * Antonyms: Vague, generic.
            * Synonyms: Detailed.
            * Quotes: “GenAI isn't specific or …”, “GenAI offered specific answers unlike my vague teachers”
        b. In-depth: Provides more in-depth or nuanced information.
            * Synonyms: nuanced, deep.
            * Antonyms: superficial, broad.
            * Quote: “GENAI does not have the nuance, it is broad”
        c. Understandable: Makes information digestible, comprehendible.
            * Quote: “GenAI is great for summarizing content and dividing large amounts of material into digestible bullet points.”
        d. Relevance: Provides information that is relevant or irrelevant. It is /not on the same topic.
            * Quote: “AI always gives irrelevant and incomplete answers”
        e. Contextualised: Provides information that is adapted for the context of an assignment, rubrics, discipline or class. (Personalised/tailored). Is not about knowing the person (if so, it is a relational code).
            * Quotes: “Teacher feedback tended to be more deep and considering the context”, “teacher feedback is specific to their marking requirements”
        f. Utility: Provides useful or helpful information.
            * Synonyms: usefulness, helpfulness.
            * Quote: “not very helpful”
        g. Reliable: Provides information that is accurate, inaccurate, precise, reliable, unreliable, trustworthy.
            * Quote: “GenAI can be unreliable when providing feedback for scientific writings”
        h. Objective: Is objective in its feedback, in its judgment, in its evaluation.
            * Antonyms: Subjective, Biased.
            * Quotes: "There may be subjective bias in teacher given feedback whereas AI is more objective and factual.”, "Teacher's feedback is always given, no matter the mark, there is always an improvement. However, I have found it to be dependent on the marker and their opinions. GenAI is objective, through it's understanding of the internet." (G2271), "Since GenAI has no bias or preference, it purely gives feedback on what it reads." (G1358), "The teacher could be inconsistent and interpret the criteria in their own way due to misreading and being poorly informed maybe due to lack of time" (G1611)
        * Tone
        i. Positivity: Makes positive statements, people pleaser.
            * Quotes: “it is also a people pleaser”, “GenAI also tends to be a ‘yes man’ or otherwise reflect the opinion of the person using it”
        j. Negativity: Makes edits or negative, insulting statements.
            * Quotes: “GenAI feedback was always polite and constructive, but unfortunately I can't say the same about all teacher/staff feedback.”, “whereas my teacher will have edits to my work.”
    4. Feeling
        a. Positive: Makes me feel a positive feeling.
            * Synonyms: Encouraging, motivating.
            * Quote: “AI made me feel good”
        b. Negative: Makes me feel a negative feeling.
            * Synonyms: Frustrating.
            * Quotes: “teachers make me feel bad and dumb”, “feedback with harsh, unhelpful wording from teachers/staff, this also had a greater impact on me emotionally.”
        c. No impact: Makes me feel nothing or indifferent. Has no impact on my emotions or feelings.
            * Quotes: “GENAI makes me feel nothing. I know it is AI working off Machine learning based on what it has learnt before.”, "Gen AI has no impact on my feelings as it’s not a person", "chatgpt is emotionless so it could say it was awful and i wouldn't care but i'd prefer not to hear that from a teacher"
    5. Relational
        * Personal: Is supportive. Concerns relational type statements (e.g. the “teacher knows me”’). Closer/ more-distant or embodied (e.g. “I like being face to face with my teacher”). The actor (teacher or AI) is trustworthy. Some statements are about AI being more in tune, more personalised, personal to tastes in language, communication and "learning styles".
            * Quotes: “GenAI can’t have a person-to-person conversation with me”, "GenAI: Offers highly personalized learning experiences, adapting to your specific questions, pace, and learning style"
        * Risky: Makes me vulnerable to their judgement, causes me shame. Possibly related to power, position, self.
            * Synonyms: Implies personal consequence, at risk.
            * Antonyms: non-judgemental, no shame, preserves self, prevents vulnerability.
            * Quotes: “I have severe anxiety and issues with asking and receive help and feedback especially in person so having AI/LLM's provide text responses help”, "AI provides a safe space to ask any and all questions that I may be too timid to ask in person. It helps me dive deeper into concepts that I struggle with", "AI made me feel more comfortable to learn compared to the teachers...", "I sometimes feel reluctant to go to the teacher with questions, meanwhile I dont feel this reluctance when going to AI."
        * Expert: Has a position of knowledge or expertise.
            * Quote: “They're also more knowledgeable in terms of the subject”
    6. Value
        * Importance: Is important or valuable.
            * Quotes: “teacher is the most important point of learning”, "It didn't impact my learning in any significant way as I always weighed the feedback from my tutors and lecturers above an AI.", "Nothing really changed, i still value teacher feedback more and ive never directly implemented a gen ai suggestion/feedback."
        
    
# Reason for NOT using GenAI:

    1. Processes
        * Access
        g. Unaware: The respondent didn't know they could or how to use GenAI, or didn't think of using it.
            * Quotes: "Didn’t think to", "I did not know that it could do that.", "It has never occurred to me but I think it's a great idea", "Its hard to use."
        * Effort
        h. Effortful: Too much effort to use or learn how to use AI; don't have time.
            * Quotes: "It is more effort than it's worth to program it to understand the confines and requirements of the work/project.", "I'm usually in a rush to submit work and don't have time."
        i. Off-loading: Off-loads cognitive effort or the experience of learning, which is not desirable.
            * Quotes: "I'm paying substantial fees to undertake this course, I figure if I can't push myself to do the entirety of the work myself and mam relying on AI, then I am not gaining the skills I need to enter the workforce.", "Because I need to learn new things with some effort. Using GenAI for work is something like I'm not using my calibre to do the work. So in order to be part of this competitive world I should hold and thrive my self with my effort.", "It is because I want to use my own intelligence to learn something new instead of depending on AI and only get marks without knowing why."

    2. Information
        * Quality
        a. Specificity: Provides specific information.
            * Antonyms: Vague, generic.
            * Synonyms: Detailed.
            * Quotes: "GenAI isn't specific or …", "GenAI offered specific answers unlike my vague teachers"
        b. In-depth: Respondent finds GenAI lacks depth.
            * Quotes: "I personally believe that GenAI lacks the nuance that would allow for effective feedback on a task."
        c. Contextualised: Information provided lacks the context of an assignment, rubrics, discipline, or class.
            * Quotes: "It does not know the specifics of the task I am doing", "GenAI will not be the ones grading me. Furthermore, GenAI likely doesn't know the rubric for grading"
        d. Trustworthy: Respondent does not trust AI or its outputs. AI is not reliable, competent, or able.
            * Quotes: "I do not trust that it would be accurate", "I would not trust GenAI to give me the correct answer or feedback", "I do not trust the americanisation of Gen AI datasets", "It's a machine, it doesn't know how to give feedback"
        e. Utility: Provides useful or helpful information (Synonyms: usefulness, helpfulness).
            * Quotes: "not very helpful"
        * Tone
        i. Positivity: Makes positive statements, "people pleaser."
            * Quotes: "it is also a people pleaser", "GenAI also tends to be a ‘yes man’ or otherwise reflect the opinion of the person using it"
    5. Relational
        a. Personal: Respondent values relationship or closer and embodied feedback. Respondent values the role of humans or connection with humans.
            * Quotes: "GenAI [can't] have a person-to-person conversation with me.", "feedback should comes from professors", "For feedback I prefer to have a human give feedback", "GenAI lacks personal emotion, I believe in its professionalism but not its humanity."
        c. Expert: Has a position of knowledge or expertise.
            * Quotes: "They're also more knowledgeable in terms of the subject"
    6. Value
        b. Preference: Respondent preferred not using GenAI without specifying further reasons or expressing dislike.
            * Quotes: "I've realized it writes in a way that I do not like.", "I just didn't, I don't really like using GenAI"
        c. Need: Does not want or need feedback. Does not need feedback from AI (own skills, or others are sufficient).
            * Quotes: "It was not necessary to use, all the tools that I needed were there", "I usually have ample opportunities for supervisors or tutors to provide feedback on my work, so I usually do not feel the need to request feedback from GenAI"
        d. Unsustainable: Respondent finds GenAI is not environmentally sustainable.
            * Quotes: "I oppose the use of GenAI on environmental grounds"
        e. Privacy: Respondent reports concern regarding privacy.
            * Quotes: "I do not want my work added into GenAI datasets"
        f. Integrity: Respondent is committed to the originality of their own work and ethical behavior in all academic work.
            * Quotes: "I want to avoid having it influence my assignments.", "I don't want to compromise the academic integrity of original ideas.", "Not sure if it is cheating", "I'm afraid of breaking the rules.", "I always assumed the use of Gen AI was associated with risks of facing a rejection of my work so I have tried to steer clear from using it at all.", "it's not allowed"


    """


"""
# Coding strategy is critical because it establishes exactly how the LLM should approach the 
# analysis of short, open-ended student responses. 
# It prevents over-interpretation, enforces a reproducible method, and handles edge cases like 
# unintelligible responses. Following GROPROE, this clarity is necessary for reliability.
"""
CODING_STRATEGY = """
When coding student responses:

1. **Surface-Level Understanding**  
   - Do not over-interpret student responses.  
   - Work from the most direct, surface-level meaning of the text.  
   - Avoid reading beyond what is explicitly written.

2. **Chunk-Based Coding**  
   - Divide each response into meaningful chunks (a sentence or a significant phrase).  
   - A chunk should only be coded once, even if it could theoretically fit multiple codes.

3. **Focus on Comparison**  
   - Prioritize identifying any comparison between AI feedback and teacher feedback.  
   - Look for explicit or implicit comparative language such as *more, less, better, worse, easier, harder, similar*.  

4. **Actor + Comparator + Characteristic Logic**  
   - For each usable chunk, identify three components:  
       **Actor** – Who or what the comment refers to (A = AI, T = Teacher, B = Both).  
       **Comparator** – The relative position (e.g., more, less, similar, better, worse).  
       **Characteristic** – The feedback property being described (e.g., clarity, accuracy, relevance, timeliness, emotional safety).  
   - A valid coded chunk must include all three components.  

5. **Handling Unusable Responses**  
   - If a response is unintelligible or does not appear to answer the question, code it as **UNUSABLE**.  
   - If a chunk is missing **any one** of the three required components (Actor, Comparator, Characteristic), code it as **UNUSABLE**.  
   - Example of unusable: "It was better" – we don’t know *which actor* it refers to.  

6. **Special Case: Question 43 (Reason for Not Using GenAI)**  
   - For responses to Question 43, the focus is on identifying **clear reasons** why students did not use AI.  
   - If no clear reason is provided, or the response is unintelligible, mark it as **UNUSABLE**.

This coding strategy ensures the LLM produces systematic, reproducible coding results while avoiding over-coding or speculative interpretation.
"""


# Output format is important because it tells the LLM exactly how to present the coding results

# in a structured way that is machine-readable and easy to evaluate.


OUTPUT_FORMAT = """
For each student reflection, output as a JSON object with:
{
"reflection_text": "...",
"codes": [
{
"category": "EXPERIENCING EMOTIONS",
"sub_category": "Academic Belonging",
"indicators": ["self-efficacy", "feeling challenged"],
"quote": "...",
"reasoning": "Explain why this code applies"
}
]
}
"""



# Instructions are important to control LLM behavior:

# - Ask it to apply all relevant codes

# - Encourage reasoning (Chain-of-Thought)

# - Avoid assumptions outside the context


LLM_INSTRUCTIONS = """

Read the reflection carefully.

Identify all segments that express affective engagement.

Match each segment to one or more codes from the coding framework.

Provide reasoning for each code assignment.

Follow the output format strictly.

Do not invent codes or interpretations outside the given framework.
"""
```
