## 2. Agent Set UP

````python
# Utility functions

def find_similar_quote(quote: str, doc: str, similarity_threshold: float = 0.7, margin: int = 40) -> tuple[bool, str]:
    """
    Find a substring in doc that is most similar to the quote and starts/ends with same letter.
    Uses optimized similarity comparison based on forward score thresholds.
    Args:
        quote (str): The quote to search for.
        doc (str): The document content.
        similarity_threshold (float): Minimum similarity score.
        margin (int): Allowed margin of error for length of matched substrings.
    Returns:
        tuple[bool, str]: (True/False, matched_text)
    """
    if not quote.strip():
        #print("[find_similar_quote] Empty quote provided.")
        return False, ""
    
    #print(f"\n[find_similar_quote] Looking for: '{quote}'")

    # Remove trailing punctuation marks from quote before processing
    import string
    quote_cleaned = quote.strip().rstrip(string.punctuation)
    quote_norm = quote_cleaned.lower()
    doc_norm = doc.lower()

    first = quote_norm[0]
    last = quote_norm[-1]
    q_len = len(quote_norm)

    best_score = 0
    best_match = ""

    for i in range(len(doc_norm)):
        for l in range(q_len - margin, q_len + margin + 1):
            if l < 5:
                continue
            j = i + l
            if j > len(doc_norm):
                continue
            candidate = doc_norm[i:j]
            if candidate[0] == first and candidate[-1] == last:
                # Forward: quote -> candidate
                #print(f"Comparing to: '{candidate}'")
                forward_score = SequenceMatcher(None, quote_norm, candidate).ratio()
                
                # Optimized scoring based on forward score thresholds
                if forward_score > 0.8:
                    # High confidence match, use forward score directly
                    final_score = forward_score
                elif forward_score > 0.4:
                    # Medium confidence, calculate backward score and average
                    backward_score = SequenceMatcher(None, candidate, quote_norm).ratio()
                    final_score = (forward_score + backward_score) / 2
                else:
                    # Low confidence, skip this candidate
                    continue
                
                #if final_score > 0.5:
                    #print(f"Compared to: {candidate}")
                # Uncomment below line if you want to see every comparison (can be a lot!)
                # print(f"  Comparing to: '{doc[i:j]}' | Forward: {forward_score:.3f}, Final: {final_score:.3f}")
                if final_score > best_score:
                    best_score = final_score
                    best_match = doc[i:j]
                    #print(f"  [NEW BEST] '{best_match}' | Final Score: {best_score:.3f}")
    
    if best_score >= similarity_threshold:
        #print(f"[find_similar_quote] Best match: '{best_match}' | Final Score: {best_score:.3f}\n")
        return True, best_match
    else:
        #print(f"[find_similar_quote] No suitable match found (best score: {best_score:.3f})")
        #print(f"[find_similar_quote] Best candidate: '{best_match}' | Score: {best_score:.3f}\n")
        return False, ""
    
def extract_json_obj(text: str) -> Optional[Dict[str, Any]]:
    """
    Extract the first top-level JSON object from arbitrary LLM text.
    Returns None if nothing parseable is found.
    """
    first = text.find('{')
    last = text.rfind('}')
    if first == -1 or last == -1 or last < first:
        return None
    snippet = text[first:last+1]
    try:
        return json.loads(snippet)
    except json.JSONDecodeError:
        # fallback: strip code fences / markdown
        cleaned = re.sub(r"^```.*?$|```", "", text, flags=re.MULTILINE|re.DOTALL)
        first = cleaned.find('{'); last = cleaned.rfind('}')
        if first == -1 or last == -1 or last < first:
            return None
        try:
            return json.loads(cleaned[first:last+1])
        except json.JSONDecodeError:
            return None

def invoke_with_timeout(messages, timeout: int = 300, retries: int = 3, backoff: float = 1.8):
    """
    Call llm.invoke(messages) with a hard timeout and simple retries.
    Raises the last exception if all attempts fail.
    """
    last_exc = None
    for attempt in range(1, retries + 1):
        try:
            with ThreadPoolExecutor(max_workers=1) as ex:
                fut = ex.submit(lambda: llm.invoke(messages))
                return fut.result(timeout=timeout)
        except TimeoutError as e:
            last_exc = e
            print(f"[ANALYZE] ⏱️ Timeout on attempt {attempt}/{retries} (>{timeout}s). Retrying…")
        except Exception as e:
            last_exc = e
            # Common: 429 rate limit / transient network errors
            print(f"[ANALYZE] ⚠️ Error on attempt {attempt}/{retries}: {e}. Retrying…")

        # backoff + jitter
        sleep_s = (backoff ** (attempt - 1)) + random.random()
        time.sleep(sleep_s)

    # Out of retries
    raise last_exc
````

````python
# =============================================================================
# 1. DATA MODELS AND STATE MANAGEMENT
# =============================================================================

# Define the state structure for our agent
class AgentState(TypedDict):
    file_path: str
    file_content: str
    processed_content: str
    analysis_result: Dict[str, Any]          # last iteration's result
    iteration_results: List[Dict[str, Any]]  # all iterations
    error: str
    iteration: int                           # current iteration (0-based)
    total_iterations: int                        # e.g., 2

# Define the output model for structured results
class Quote(BaseModel):
    quote: str = Field(description="Exact quote from the transcript")
    reasoning: str = Field(description="Explanation of why this code applies")
    validation_status: str = Field(description="Validation status: 'valid', 'invalid', or 'pending'")
    similarity_score: float = Field(description="Similarity score with matched text in document", default=0.0)
    matched_text: str = Field(description="Actual matched text from document", default="")

class CodeResult(BaseModel):
    category: str = Field(description="Main category from coding framework")
    theme: str = Field(description="Theme within the category")
    sub_theme: str = Field(description="Sub-theme within the theme")
    code: str = Field(description="Specific code applied")
    quotes: List[Quote] = Field(description="List of supporting quotes")

class AnalysisOutput(BaseModel):
    codes: List[CodeResult] = Field(description="List of all codes found in the transcript")

# =============================================================================
# 2. UTILITY FUNCTIONS
# =============================================================================

def read_transcript_file(file_path: str) -> str:
    """Read and clean transcript file content"""
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
        
        # Clean the content - remove timestamps and speaker labels for analysis
        lines = content.split('\n')
        cleaned_lines = []
        
        for line in lines:
            # Skip timestamp lines (format: 00:00:00.000 --> 00:00:00.000)
            if re.match(r'^\d{2}:\d{2}:\d{2}\.\d{3}\s*-->\s*\d{2}:\d{2}:\d{2}\.\d{3}$', line.strip()):
                continue
            
            # Skip line numbers
            if re.match(r'^\d+$', line.strip()):
                continue
            
            # Skip empty lines
            if not line.strip():
                continue
            
            # Keep meaningful content
            cleaned_lines.append(line.strip())
        
        return '\n'.join(cleaned_lines)
    
    except Exception as e:
        raise Exception(f"Error reading file {file_path}: {str(e)}")

def extract_json_obj(response_text: str) -> Dict[str, Any]:
    """Extract JSON object from response text"""
    try:
        # Find JSON in the response
        json_start = response_text.find('{')
        json_end = response_text.rfind('}') + 1
        
        if json_start != -1 and json_end != 0:
            json_str = response_text[json_start:json_end]
            return json.loads(json_str)
        return None
    except json.JSONDecodeError:
        return None
    
def validate_quote_tool(quote: str, document_content: str) -> dict:
    """
    Tool function to validate if a quote exists in the document content.
    Returns validation result with similarity score and matched text.
    """
    from difflib import SequenceMatcher
    
    # Use the existing find_similar_quote function
    is_valid, matched_text = find_similar_quote(quote, document_content)
    
    # Calculate similarity score for reporting
    if matched_text:
        similarity_score = SequenceMatcher(None, quote.lower(), matched_text.lower()).ratio()
    else:
        similarity_score = 0.0
    
    return {
        "is_valid": is_valid,
        "original_quote": quote,
        "matched_text": matched_text,
        "similarity_score": similarity_score,
        "validation_status": "valid" if is_valid else "invalid"
    }

# =============================================================================
# 3. WORKFLOW NODE FUNCTIONS (in execution order)
# =============================================================================

def process_transcript_file(state: AgentState) -> AgentState:
    """Step 1: Process and clean the transcript file"""
    try:
        print(f"[PROCESS] Reading and cleaning: {state['file_path']}")
        content = read_transcript_file(state['file_path'])
        state['file_content'] = content
        state['processed_content'] = content
        print(f"[PROCESS] Cleaned transcript length: {len(content.split())} words")
    except Exception as e:
        state['error'] = f"Processing failed: {str(e)}"
    return state


def analyze_one_iteration(state: AgentState) -> AgentState:
    try:
        display_iter = state['iteration'] + 1
        print(f"[ANALYZE] Iteration {display_iter}/{state['total_iterations']}")

        analysis_prompt = f"""
{initial_prompt}

You are running iteration {display_iter} of {state['total_iterations']}.
Here is the transcript:

{state['processed_content']}

Return ONLY valid JSON in the agreed schema.
"""

        print("[ANALYZE] → sending prompt to LLM")
        response = invoke_with_timeout(
            [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": analysis_prompt}
            ],
            timeout=300,
            retries=3,
            backoff=1.8
        )
        print("[ANALYZE] ← response received")

        response_text = getattr(response, "content", str(response))
        result = extract_json_obj(response_text)
        if result is None:
            state['error'] = "No valid JSON found in response"
            print("[ANALYZE] ❌ No valid JSON found")
            return state

        state['analysis_result'] = result
        state['iteration_results'].append(result)
        print(f"[ANALYZE] ✅ Codes found: {len(result.get('codes', []))}")

        # increment after a successful iteration
        state['iteration'] += 1

    except Exception as e:
        state['error'] = f"Analysis failed: {str(e)}"
        print(f"[ANALYZE] ❌ Failure: {e}")
    return state



def validate_quotes(state: AgentState) -> AgentState:
    """Step 2.5: Validate all quotes found in the analysis"""
    try:
        print("[VALIDATE] Validating quotes from analysis")
        
        if not state.get('analysis_result'):
            state['error'] = "No analysis result to validate"
            return state
        
        codes = state['analysis_result'].get('codes', [])
        validated_codes = []
        
        for code in codes:
            validated_quotes = []
            for quote_obj in code.get('quotes', []):
                quote_text = quote_obj.get('quote', '')
                
                # Use the validation tool
                validation_result = validate_quote_tool(quote_text, state['processed_content'])
                
                # Update quote with validation info
                validated_quote = {
                    **quote_obj,
                    'validation_status': validation_result['validation_status'],
                    'similarity_score': validation_result['similarity_score'],
                    'matched_text': validation_result['matched_text']
                }
                validated_quotes.append(validated_quote)
            
            # Update code with validated quotes
            validated_code = {**code, 'quotes': validated_quotes}
            validated_codes.append(validated_code)
        
        # Update state with validated results
        state['analysis_result']['codes'] = validated_codes
        
        # Count validation results
        total_quotes = sum(len(code.get('quotes', [])) for code in validated_codes)
        valid_quotes = sum(
            sum(1 for q in code.get('quotes', []) if q.get('validation_status') == 'valid')
            for code in validated_codes
        )
        
        print(f"[VALIDATE] ✅ Validated {total_quotes} quotes: {valid_quotes} valid, {total_quotes - valid_quotes} invalid")
        
    except Exception as e:
        state['error'] = f"Quote validation failed: {str(e)}"
    
    return state

def review_iterations(state: AgentState, master_csv_path: str = "./coding_framework.csv") -> AgentState:
    """
    Step 3: Compare all iterations with allowed codes from master CSV and print existence checks.
    No saving — only prints results.
    """
    try:
        print("[REVIEW] Checking codes against master list")

        # Load master list
        master_df = pd.read_csv(master_csv_path)

        def _norm(s):
            return "" if pd.isna(s) else str(s).strip().lower()

        has_category = "category" in (c.lower() for c in master_df.columns)

        # Normalize column names
        master_df.columns = [c.lower() for c in master_df.columns]

        allowed_themes     = set(_norm(x) for x in master_df["theme"].unique())
        allowed_sub_themes = set(_norm(x) for x in master_df["sub_theme"].unique())
        allowed_codes      = set(_norm(x) for x in master_df["code"].unique())

        if has_category:
            allowed_categories = set(_norm(x) for x in master_df["category"].unique())
            allowed_quads = {
                (_norm(r["category"]), _norm(r["theme"]), _norm(r["sub_theme"]), _norm(r["code"]))
                for _, r in master_df.iterrows()
            }
        else:
            allowed_quads = None

        allowed_triples = {
            (_norm(r["theme"]), _norm(r["sub_theme"]), _norm(r["code"]))
            for _, r in master_df.iterrows()
        }

        # Iterate through agent results
        iters = state.get("iteration_results", [])
        for i, res in enumerate(iters, start=1):
            print(f"\n--- Iteration {i} ---")
            codes = (res or {}).get("codes", [])
            for c in codes:
                cat   = _norm(c.get("category", ""))
                theme = _norm(c.get("theme", ""))
                sub   = _norm(c.get("sub_theme", ""))
                code  = _norm(c.get("code", ""))

                exists_category = True if not has_category else (cat in allowed_categories)
                exists_theme    = theme in allowed_themes
                exists_sub      = sub in allowed_sub_themes
                exists_code     = code in allowed_codes

                if has_category:
                    exists_combo = (cat, theme, sub, code) in allowed_quads
                else:
                    exists_combo = (theme, sub, code) in allowed_triples

                print(
                    f"Category: {c.get('category', '')} "
                    f"Theme: {c.get('theme', '')} "
                    f"Sub-theme: {c.get('sub_theme', '')} "
                    f"Code: {c.get('code', '')} "
                    f"=> Exists? Category={exists_category}, Theme={exists_theme}, "
                    f"Sub={exists_sub}, Code={exists_code}, Combo={exists_combo}"
                )

    except Exception as e:
        state["error"] = f"Reviewer failed: {str(e)}"
        print(f"[REVIEW] ❌ {e}")
    return state



def save_analysis_result(state: AgentState) -> AgentState:
    """Step 4: Save all analysis results"""
    try:
        base_name = os.path.splitext(os.path.basename(state['file_path']))[0]
        base_dir  = f"./analysis_results/{base_name}"
        os.makedirs(base_dir, exist_ok=True)

        for i, res in enumerate(state.get('iteration_results', []), start=1):
            file_path = os.path.join(base_dir, f"iter_{i}.json")
            with open(file_path, "w", encoding="utf-8") as f:
                json.dump(res, f, indent=2, ensure_ascii=False)
            print(f"[SAVE] Iteration {i} saved: {file_path}")

        combined = {
            "file": base_name,
            "total_iterations": state.get("total_iterations", 0),
            "iterations": state.get("iteration_results", []),
            "last_result": state.get("analysis_result", {})
        }
        combined_path = os.path.join(base_dir, "combined.json")
        with open(combined_path, "w", encoding="utf-8") as f:
            json.dump(combined, f, indent=2, ensure_ascii=False)
        print(f"[SAVE] Combined file saved: {combined_path}")

    except Exception as e:
        state['error'] = f"Failed to save result: {str(e)}"
    return state


# =============================================================================
# 4. WORKFLOW CONTROL FUNCTIONS
# =============================================================================

def should_continue(state: AgentState) -> str:
    """Decide next step without mutating state."""
    if state.get('error'):
        print("[DEBUG] Error detected → save")
        return "save"

    it = state.get('iteration', 0)              # already incremented by analyze node
    total = state.get('total_iterations', 1)

    # keep prints lightweight
    print(f"[DEBUG] Progress: {it}/{total} iterations completed")

    if it < total:
        return "loop"       # run analyze_iteration again
    else:
        return "review"     # iterations finished → reviewer

def create_analysis_workflow():
    """Create the workflow graph with quote validation"""
    workflow = StateGraph(AgentState)

    workflow.add_node("process_file", process_transcript_file)
    workflow.add_node("analyze_iteration", analyze_one_iteration)
    workflow.add_node("validate_quotes", validate_quotes)
    workflow.add_node("review_iterations", review_iterations)
    workflow.add_node("save_result", save_analysis_result)

    workflow.add_edge(START, "process_file")
    workflow.add_edge("process_file", "analyze_iteration")
    workflow.add_edge("analyze_iteration", "validate_quotes")

    workflow.add_conditional_edges(
        "validate_quotes",
        should_continue,
        {
            "loop": "analyze_iteration",  # This will run the next iteration
            "review": "review_iterations",
            "save": "save_result",
        }
    )

    workflow.add_edge("review_iterations", "save_result")
    workflow.add_edge("save_result", END)
    return workflow.compile()

# app = create_analysis_workflow()

# Image(app.get_graph(xray=1).draw_mermaid_png())
````

````python
def _df_from_analysis_result(analysis_result):
    """Convert analysis result to DataFrame for display"""
    rows = []
    if analysis_result and 'codes' in analysis_result:
        for c in analysis_result['codes']:
            rows.append({
                "theme": c.get("theme", ""),
                "sub_theme": c.get("sub_theme", ""),
                "code": c.get("code", ""),
                "quotes_count": len(c.get("quotes", []))
            })
    return pd.DataFrame(rows, columns=["theme", "sub_theme", "code", "quotes_count"])

def test_single_file_nb(
    filename: str,
    transcripts_dir: str = "./temp_survey",
    total_iterations: int = 2,
    show_quotes: bool = False
):
    """Run the workflow on ONE file and print results inline."""
    file_path = os.path.join(transcripts_dir, filename)
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        return
    
    display(Markdown(f"### ▶️ Testing file: `{filename}`"))
    workflow = create_analysis_workflow()

    initial_state = {
        "file_path": file_path,
        "file_content": "",
        "processed_content": "",
        "analysis_result": {},
        "iteration_results": [],  # Add this
        "error": "",
        "iteration": 0,          # Add this
        "total_iterations": total_iterations  # Add this
    }

    try:
        final_state = workflow.invoke(initial_state)
    except Exception as e:
        print(f"Workflow failed: {e}")
        return

    if final_state.get("error"):
        print("❌ Error:", final_state["error"])
        return

    # Summary
    print("✅ Analysis completed successfully")
    print("Iterations:", final_state.get("total_iterations", 0))

    # Display results
    if final_state.get("analysis_result"):
        df = _df_from_analysis_result(final_state["analysis_result"])
        display(Markdown("**Themes / Sub-themes / Codes:**"))
        display(df)
        
        # Show quotes if requested
        if show_quotes and final_state.get("analysis_result", {}).get("codes"):
            print("\nQuotes by code:")
            for c in final_state["analysis_result"]["codes"]:
                code = c.get("code", "")
                print(f"\n--- {code} ---")
                for q in c.get("quotes", []):
                    print(f"Quote: {q.get('quote','').strip()}")
                    print(f"Reasoning: {q.get('reasoning','').strip()}")
    else:
        print("No analysis results found")
from IPython.display import Markdown, display
import os, time
import pandas as pd

def _df_from_analysis_result(analysis_result):
    """Convert analysis result to a small, readable DataFrame."""
    rows = []
    if analysis_result and 'codes' in analysis_result:
        for c in analysis_result['codes']:
            rows.append({
                "category":  c.get("category", ""),
                "theme":     c.get("theme", ""),
                "sub_theme": c.get("sub_theme", ""),
                "code":      c.get("code", ""),
                "quotes_count": len(c.get("quotes", []))
            })
    return pd.DataFrame(rows, columns=["category", "theme", "sub_theme", "code", "quotes_count"])

def test_all_files_nb(
    transcripts_dir: str = "./temp_survey",
    total_iterations: int = 2,
    show_quotes: bool = False,
    show_validation: bool = True
):
    """
    Run the workflow on ALL .txt files in a folder and print results inline.
    - total_iterations: how many analyze/validate loops to run per file
    - show_quotes: print quotes + reasoning per code
    - show_validation: print the existence checks produced by review_iterations (if available)
    Returns a pandas.DataFrame with one row per (file, code).
    """
    if not os.path.exists(transcripts_dir):
        print(f"Folder not found: {transcripts_dir}")
        return pd.DataFrame()

    files = sorted([f for f in os.listdir(transcripts_dir) if f.endswith(".txt")])
    if not files:
        print(f"No .txt files found in {transcripts_dir}")
        return pd.DataFrame()

    display(Markdown(f"## Batch run: {len(files)} file(s) from `{transcripts_dir}`"))
    workflow = create_analysis_workflow()

    batch_rows = []

    for idx, filename in enumerate(files, start=1):
        file_path = os.path.join(transcripts_dir, filename)
        display(Markdown(f"### ▶️ `{filename}`  _(file {idx}/{len(files)})_"))
        t0 = time.time()

        initial_state = {
            "file_path": file_path,
            "file_content": "",
            "processed_content": "",
            "analysis_result": {},
            "iteration_results": [],
            "error": "",
            "iteration": 0,
            "total_iterations": total_iterations
        }

        try:
            final_state = workflow.invoke(initial_state)
        except Exception as e:
            print(f"❌ Workflow failed: {e}")
            continue

        if final_state.get("error"):
            print("❌ Error:", final_state["error"])
            continue

        # Per‑file summary table
        if final_state.get("analysis_result"):
            df = _df_from_analysis_result(final_state["analysis_result"])
            if len(df) == 0:
                print("No codes found.")
            else:
                display(df)

                # Accumulate rows for batch summary
                for _, r in df.iterrows():
                    batch_rows.append({
                        "file": filename,
                        "category":  r["category"],
                        "theme":     r["theme"],
                        "sub_theme": r["sub_theme"],
                        "code":      r["code"],
                        "quotes_count": int(r["quotes_count"])
                    })
        else:
            print("No analysis results found.")

        # Optional: print quotes + reasoning
        if show_quotes and final_state.get("analysis_result", {}).get("codes"):
            print("\nQuotes by code:")
            for c in final_state["analysis_result"]["codes"]:
                code = c.get("code", "")
                print(f"\n— {code} —")
                for q in c.get("quotes", []):
                    qt = (q.get('quote','') or '').strip()
                    rs = (q.get('reasoning','') or '').strip()
                    print(f"  • Quote: {qt}")
                    if rs:
                        print(f"    Reasoning: {rs}")

        # Optional: print validation checks made in review_iterations
        if show_validation and final_state.get("validation_preview"):
            print("\nValidation checks (first 10 rows):")
            val_preview = final_state["validation_preview"]
            # pretty print
            for row in val_preview:
                print(
                    f"  [{row.get('iteration','?')}] "
                    f"category='{row.get('category','')}', "
                    f"theme='{row.get('theme','')}', "
                    f"sub_theme='{row.get('sub_theme','')}', "
                    f"code='{row.get('code','')}' "
                    f"=> exists: cat={row.get('exists_category', True)}, "
                    f"theme={row.get('exists_theme', False)}, "
                    f"sub={row.get('exists_sub_theme', False)}, "
                    f"code={row.get('exists_code', False)}, "
                    f"combo={row.get('exists_full_combo', False)}"
                )

        dt = time.time() - t0
        print(f"✅ Done in {dt:.1f}s\n")

    # Batch summary (returned)
    batch_df = pd.DataFrame(
        batch_rows,
        columns=["file", "category", "theme", "sub_theme", "code", "quotes_count"]
    )

    if len(batch_df) > 0:
        display(Markdown("## Batch summary"))
        # quick pivot: how many codes per file
        pivot = batch_df.groupby("file")["code"].count().reset_index().rename(columns={"code":"codes_found"})
        display(pivot)

    return batch_df
````
